## Weekly Total Cleanup

```python
wklytotal.dropna
wklytotal.fillna(0, inplace = True)
wklytotal.head(9)
```

```python
# merge top row with column names
wklytotal.columns = np.append(wklytotal.iloc[0, :33].astype('int'), wklytotal.columns[33:])
wklytotal = wklytotal.iloc[1:].reset_index(drop=True)
```

```python
#drops any row that has 0's in the entire row
wklydrop = wklytotal.loc[(wklytotal!=0).any(1)]
```

```python
# reset index after dropping the dead rows
wklydrop.reset_index(drop=True, inplace=True)
```

```python
# rename the columns the easy way. next time, just import stringIO 
# needed to give placeholder names to group columns
wklydrop.columns = ['Week', 'Day', '2017', '2018', '2019','2020','2021', 
                    'Week.a','Day.a', '2017.a', '2018.a', '2019.a','2020.a','2021.a', 
                    'Week.b', 'Day.b', '2017.b', '2018.b', '2019.b','2020.b','2021.b',
                   'Week.c', 'Day.c', '2017.c', '2018.c', '2019.c','2020.c','2021.c',
                   'Week.d', 'Day.d', '2018.d', '2019.d','2020.d']
```

```python
# using for loops to rename the NaN/0 into the correct week number
# going to copy the for loop for each df because im lazy
def weekly():
    for i in range(1, len(Q1)):
        if Q1.loc[i, 'Week'] == 0:
            Q1.loc[i, 'Week'] = Q1.loc[i-1, 'Week']
        else:
            continue
    for i in range(1, len(Q2)):
        if Q2.loc[i, 'Week'] == 0:
            Q2.loc[i, 'Week'] = Q2.loc[i-1, 'Week']
        else:
            continue
    for i in range(1, len(Q3)):
        if Q3.loc[i, 'Week'] == 0:
            Q3.loc[i, 'Week'] = Q3.loc[i-1, 'Week']
        else:
            continue
    for i in range(1, len(Q4)):
        if Q4.loc[i, 'Week'] == 0:
            Q4.loc[i, 'Week'] = Q4.loc[i-1, 'Week']
        else:
            continue
```

```python
# functions to change each day name in each quarter DF. 
def Q1day():
    for i in range(0, len(Q1)):
        if Q1.loc[i, 'Day'] == 'M':
            Q1.loc[i, 'Day'] = 'Monday'
        elif Q1.loc[i, 'Day'] == 'T' and Q1.loc[i-1, 'Day'] == 'Monday':
            Q1.loc[i, 'Day'] = 'Tuesday'
        elif Q1.loc[i, 'Day'] == 'W':
            Q1.loc[i, 'Day'] = 'Wednesday'
        elif Q1.loc[i, 'Day'] == 'T' and Q1.loc[i-1, 'Day'] == 'Wednesday':
            Q1.loc[i, 'Day'] = 'Thursday'
        elif Q1.loc[i, 'Day'] == 'F':
            Q1.loc[i, 'Day'] = 'Friday'
        elif Q1.loc[i, 'Day'] == 'S' and Q1.loc[i-1, 'Day'] == 'Friday':
            Q1.loc[i, 'Day'] = 'Saturday'     
        elif Q1.loc[i, 'Day'] == 'S' and Q1.loc[i-1, 'Day'] == 'Saturday':
            Q1.loc[i, 'Day'] = 'Sunday'
        else:
            continue

def Q2day():
    for i in range(0, len(Q2)):
        if Q2.loc[i, 'Day'] == 'M':
            Q2.loc[i, 'Day'] = 'Monday'
        elif Q2.loc[i, 'Day'] == 'T' and Q2.loc[i-1, 'Day'] == 'Monday':
            Q2.loc[i, 'Day'] = 'Tuesday'
        elif Q2.loc[i, 'Day'] == 'W':
            Q2.loc[i, 'Day'] = 'Wednesday'
        elif Q2.loc[i, 'Day'] == 'T' and Q2.loc[i-1, 'Day'] == 'Wednesday':
            Q2.loc[i, 'Day'] = 'Thursday'
        elif Q2.loc[i, 'Day'] == 'F':
            Q2.loc[i, 'Day'] = 'Friday'
        elif Q2.loc[i, 'Day'] == 'S' and Q2.loc[i-1, 'Day'] == 'Friday':
            Q2.loc[i, 'Day'] = 'Saturday'     
        elif Q2.loc[i, 'Day'] == 'S' and Q2.loc[i-1, 'Day'] == 'Saturday':
            Q2.loc[i, 'Day'] = 'Sunday'
        else:
            continue

def Q3day():
    for i in range(0, len(Q3)):
        if Q3.loc[i, 'Day'] == 'M':
            Q3.loc[i, 'Day'] = 'Monday'
        elif Q3.loc[i, 'Day'] == 'T' and Q3.loc[i-1, 'Day'] == 'Monday':
            Q3.loc[i, 'Day'] = 'Tuesday'
        elif Q3.loc[i, 'Day'] == 'W':
            Q3.loc[i, 'Day'] = 'Wednesday'
        elif Q3.loc[i, 'Day'] == 'T' and Q3.loc[i-1, 'Day'] == 'Wednesday':
            Q3.loc[i, 'Day'] = 'Thursday'
        elif Q3.loc[i, 'Day'] == 'F':
            Q3.loc[i, 'Day'] = 'Friday'
        elif Q3.loc[i, 'Day'] == 'S' and Q3.loc[i-1, 'Day'] == 'Friday':
            Q3.loc[i, 'Day'] = 'Saturday'     
        elif Q3.loc[i, 'Day'] == 'S' and Q3.loc[i-1, 'Day'] == 'Saturday':
            Q3.loc[i, 'Day'] = 'Sunday'
        else:
            continue
            
def Q4day():
    for i in range(0, len(Q4)):
        if Q4.loc[i, 'Day'] == 'M':
            Q4.loc[i, 'Day'] = 'Monday'
        elif Q4.loc[i, 'Day'] == 'T' and Q4.loc[i-1, 'Day'] == 'Monday':
            Q4.loc[i, 'Day'] = 'Tuesday'
        elif Q4.loc[i, 'Day'] == 'W':
            Q4.loc[i, 'Day'] = 'Wednesday'
        elif Q4.loc[i, 'Day'] == 'T' and Q4.loc[i-1, 'Day'] == 'Wednesday':
            Q4.loc[i, 'Day'] = 'Thursday'
        elif Q4.loc[i, 'Day'] == 'F':
            Q4.loc[i, 'Day'] = 'Friday'
        elif Q4.loc[i, 'Day'] == 'S' and Q4.loc[i-1, 'Day'] == 'Friday':
            Q4.loc[i, 'Day'] = 'Saturday'     
        elif Q4.loc[i, 'Day'] == 'S' and Q4.loc[i-1, 'Day'] == 'Saturday':
            Q4.loc[i, 'Day'] = 'Sunday'
        else:
            continue    

def Q4aday():
    for i in range(0, len(Q4a)):
        if Q4a.loc[i, 'Day'] == 'M':
            Q4a.loc[i, 'Day'] = 'Monday'
        elif Q4a.loc[i, 'Day'] == 'T' and Q4a.loc[i-1, 'Day'] == 'Monday':
            Q4a.loc[i, 'Day'] = 'Tuesday'
        elif Q4a.loc[i, 'Day'] == 'W':
            Q4a.loc[i, 'Day'] = 'Wednesday'
        elif Q4a.loc[i, 'Day'] == 'T' and Q4a.loc[i-1, 'Day'] == 'Wednesday':
            Q4a.loc[i, 'Day'] = 'Thursday'
        elif Q4a.loc[i, 'Day'] == 'F':
            Q4a.loc[i, 'Day'] = 'Friday'
        elif Q4a.loc[i, 'Day'] == 'S' and Q4a.loc[i-1, 'Day'] == 'Friday':
            Q4a.loc[i, 'Day'] = 'Saturday'     
        elif Q4a.loc[i, 'Day'] == 'S' and Q4a.loc[i-1, 'Day'] == 'Saturday':
            Q4a.loc[i, 'Day'] = 'Sunday'
        else:
            continue    
```

```python
# rename quarterly to specific quarter
# catagorize quarterly as a Ttl value. This code only changes it because there arent any other 0's in column
Q1['Week'] = Q1['Week'].str.replace('Quarterly','Q1')
Q1['Day'] = Q1['Day'].astype(str)
Q1['Day'] = Q1['Day'].str.replace('Ttl','Total')
Q1['Day'] = Q1['Day'].str.replace('0','Total')
Q1day()
```
