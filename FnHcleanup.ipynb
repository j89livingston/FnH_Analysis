## Weekly Total Cleanup

```python
wklytotal.dropna
wklytotal.fillna(0, inplace = True)
wklytotal.head(9)
```

```python
# only run this once
wklytotal.columns = np.append(wklytotal.iloc[0, :33].astype('int'), wklytotal.columns[33:])
wklytotal = wklytotal.iloc[1:].reset_index(drop=True)
```

```python
#drops any row that has 0's in the entire row
wklydrop = wklytotal.loc[(wklytotal!=0).any(1)]
```

```python
# reset index after dropping the dead rows
wklydrop.reset_index(drop=True, inplace=True)
```

```python
# rename the columns the easy way. next time, just import stringIO 
# needed to give placeholder names to group columns
wklydrop.columns = ['Week', 'Day', '2017', '2018', '2019','2020','2021', 
                    'Week.a','Day.a', '2017.a', '2018.a', '2019.a','2020.a','2021.a', 
                    'Week.b', 'Day.b', '2017.b', '2018.b', '2019.b','2020.b','2021.b',
                   'Week.c', 'Day.c', '2017.c', '2018.c', '2019.c','2020.c','2021.c',
                   'Week.d', 'Day.d', '2018.d', '2019.d','2020.d']
```
